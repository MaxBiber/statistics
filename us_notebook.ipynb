```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:,.2f}'.format # Print only 2 decimal cases.
```

```python
college = pd.read_csv("College_Data.xls")
college.head()
```

```python
college = college.set_index("Unnamed: 0") # The default option 'drop=True', deletes the column
college.index.name = 'Names'
college.head()
```

```python
college.describe(include='all')
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
```

```python
# Génération des boxplots côte à côte
college.boxplot(column='Outstate', by='Private', figsize=(8, 6))

# Ajout d'un titre et affichage
plt.title('Outstate vs Private')
plt.suptitle('')  # Supprime le sous-titre automatique ajouté par pandas
plt.xlabel('Private')
plt.ylabel('Outstate')
plt.show()
```

```python
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite', figsize=(8, 6))
```

```python
# Choose quantitative variables for histograms
variables = ['Outstate', 'Apps', 'Enroll', 'Top10perc']

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot histograms for each variable with different bin counts
axs[0, 0].hist(college['Outstate'], bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Outstate Tuition')

axs[0, 1].hist(college['Apps'], bins=30, color='salmon', edgecolor='black')
axs[0, 1].set_title('Number of Applications (Apps)')

axs[1, 0].hist(college['Enroll'], bins=15, color='lightgreen', edgecolor='black')
axs[1, 0].set_title('Number of Enrollments (Enroll)')

axs[1, 1].hist(college['Top10perc'], bins=25, color='coral', edgecolor='black')
axs[1, 1].set_title('Top 10% Percentage (Top10perc)')

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```
